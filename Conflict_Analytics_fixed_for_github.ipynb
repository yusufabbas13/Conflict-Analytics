```python
from sentence_transformers import SentenceTransformer
import torch

model = SentenceTransformer('all-MiniLM-L6-v2')

embeddings = model.encode(
    df["notes"].tolist(),
    batch_size=64,
    show_progress_bar=True
)
```

```python
import numpy as np

embeddings_df = pd.DataFrame(
    embeddings,
    columns=[f"bert_{i}" for i in range(embeddings.shape[1])]
)
```

```python
embeddings_df
```

```python
#embeddings_df = pd.read_csv("/content/embeddings_df.csv")
```

```python
df = pd.concat([df.reset_index(drop=True), embeddings_df], axis=1)
```

```python
df
```

```python
df["week"] = df["event_date"].dt.to_period("W")

bert_cols = [cols for cols in df.columns if cols.startswith("bert_")]
actor_cols = [cols for cols in df.columns if cols.startswith("actor_")]
event_type_cols = [cols for cols in df.columns if cols.startswith("event_type_")]

weekly_df = df.groupby(["country", "week"]).agg(
    {
    "fatalities": "sum",
    **{col: "sum" for col in actor_cols},
    **{col: "sum" for col in event_type_cols},
    **{col: "mean" for col in bert_cols}
    }
).reset_index()
```

```python
# embeddings_df.to_csv("/content/embeddings_df.csv", index = False)
```

```python
weekly_df
```

```python
weekly_df["event_count"] = (
    weekly_df[
        [
            "event_type_Battles",
            "event_type_Explosions/Remote violence",
            "event_type_Protests",
            "event_type_Riots",
            "event_type_Strategic developments",
            "event_type_Violence against civilians"
        ]
    ].sum(axis=1)
)
```

```python
weekly_df["violent_events"] = (
    weekly_df["event_type_Battles"] +
    weekly_df["event_type_Explosions/Remote violence"] +
    weekly_df["event_type_Violence against civilians"]
)

weekly_df["violent_event_ratio"] = (
    weekly_df["violent_events"] /
    weekly_df["event_count"].replace(0,1)
)
```

```python
weekly_df["state_actor_ratio"] = (
    weekly_df["actor_State forces"] /
    weekly_df[
        [
            "actor_Civilians",
            "actor_External/Other forces",
            "actor_Identity militia",
            "actor_Political militia",
            "actor_Protesters",
            "actor_Rebel group",
            "actor_Rioters",
            "actor_State forces"
        ]
    ].sum(axis=1).replace(0,1)
)
```

```python
weekly_df["fatalities_lag2"] = (
    weekly_df.groupby("country")["fatalities"]
    .rolling(2)
    .sum()
    .reset_index(0,drop=True)
)
```

```python
weekly_df["fatalities_lag4"] = (
    weekly_df.groupby("country")["fatalities"]
    .rolling(4)
    .sum()
    .reset_index(0,drop=True)
)
```

```python
event_cols = [
    "event_type_Battles",
    "event_type_Explosions/Remote violence",
    "event_type_Protests",
    "event_type_Riots",
    "event_type_Strategic developments",
    "event_type_Violence against civilians"
]

weekly_df["event_diversity"] = (weekly_df[event_cols] > 0).sum(axis=1)
```

```python
weekly_df["future_fatalities"] = (
    weekly_df.groupby("country")["fatalities"]
    .shift(-1) +
    weekly_df.groupby("country")["fatalities"].shift(-2) +
    weekly_df.groupby("country")["fatalities"].shift(-3) +
    weekly_df.groupby("country")["fatalities"].shift(-4)
)
```

```python
weekly_df["conflict_escalation"] = (
    weekly_df["future_fatalities"] > 2 * weekly_df["fatalities"]
).astype(int)
```

```python
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier

X = weekly_df.drop(columns=["conflict_escalation","country","week"])
y = weekly_df["conflict_escalation"]

X_train,X_test,y_train,y_test = train_test_split(
    X,y,test_size=0.2,shuffle=False
)

model = XGBClassifier(
    n_estimators=200,
    max_depth=6,
    learning_rate=0.05
)

model.fit(X_train,y_train)
```

```python
probs = model.predict_proba(X_test)[:,1]
```

```python
risk_index = probs * 100
```
